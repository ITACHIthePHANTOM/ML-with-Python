```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error

X,Y = make_regression(n_samples=100, n_features=10, noise=5, random_state=42)
X
```

```python
print(X.shape + Y.shape)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
X_train
```

```python
display(X_train.mean())
display(X_train.max())
display(X_train.min())
display(X_train.std())
display(X_test.mean())
display(X_test.max())
display(X_test.min())
display(X_test.std())
```

```python
pca = PCA()
X_train_pca = pca.fit_transform(X_train)

X_train_pca.shape
```

```python
X_train_pca
```

```python
X_test.shape
```

```python
evr = pca.explained_variance_ratio_
evr
```

```python
#10 values of evr for 10 columns. 1st value is PC1. so PC1=0.148 means PC1 is capturing 14% of the dataset.
#but we're not understanding much from here, doesn't make much sense. So we'll do a cummulative summation.

evr = np.cumsum(evr)
evr
```

```python
#So now we'll see how many of the PCAs will cover 90% of the data set. That many columns we need.
#PCA is nothing, but the new columns we're creating. So here we see, that we need 9 columns out of the 10 columns.
# i.e, reducing dimensionality by 1.
# we can show the same thing by plotting as well.

plt.plot(range(1, len(evr)+1), evr, marker='o')
plt.xlabel("Number of PC")
plt.ylabel("evr")
plt.show()
```

```python
#now since we know the no. of PCAs, we can again do....

pca = PCA(n_components=9)
X_train_pca = pca.fit_transform(X_train)

X_train_pca.shape
```

```python
#notice how columns reduced from 10 to 9
from sklearn.svm import SVR
svr_model = SVR(kernel='linear')
X_test = pca.transform(X_test)
```

```python
svr_model.fit(X_train, Y_train)
```

```python
Y_pred = svr_model.predict(X_test)
Y_pred
```
