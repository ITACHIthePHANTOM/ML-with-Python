```python
import pandas as pd
import yfinance as yf

def fetch_stocks_data(name, start_date, end_date):
    stock_data = yf.download(name, start_date, end_date)
    return stock_data

stock_data = fetch_stocks_data("AAPL","2023-01-01","2024-01-01")
stock_data
```

```python
correlation=stock_data.corr()
correlation
```

```python
stock_data.drop(['Volume'], axis=1)
```

```python
stock_data
```

```python
#So let's take the first 4 columns as input and drop the last column-Adj Close.

X=stock_data.drop(['Adj Close'],axis=1)
X
```

```python
Y=stock_data['Adj Close']     #taking Y as Output as Adj Close
Y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y, test_size=0.2,random_state=42)
X_train
```

```python
from IPython.display import display_html
display(X_train)
display(X_test)
display(Y_train)
display(Y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)    #fit_transform() method in StandardScaler is used to scale data so that 
                                     #it has a mean of 0 and a standard deviation of 1.
X_train
```

```python
display(X_train.mean()) # the value below, 3.5*e^(-16) comes close to 0 only!!!
display(X_train.max())
display(X_train.min())
```

```python
X_test=sc.transform(X_test) #no need to do fit_transform again. Only transform will do, bcz already fit once with X_train.
X_test
```

```python
display(X_test.mean()) # the value below, for mean comes close to 0 only!!!
display(X_test.max())    #again we can see, how mean comes almost middle to max and min.
display(X_test.min())
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,Y_train)  #we need to fit X and Y
```

```python
lr.intercept_   #value of c , i.e. y-intercept
```

```python
lr.coef_       #value of m or slope. Note we got 5 values, since there were 5 columns. One value for each column 
```

```python
Y_pred=lr.predict(X_test)
Y_pred
```

```python
lr.score(X_test,Y_pred)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mse=mean_squared_error(Y_test,Y_pred)
mse
```
